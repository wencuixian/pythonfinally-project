### 分析一
从世界地图来看，世界各国的GDP颜色深浅差异较大，其中一些发达国家，例如美国，西欧，北欧等国家的GDP已经远远的超过了一些发展中国家，例如非洲大陆、南美洲。而下面的滑动图可以将效果展示的更加明显，因为所有的国家都在同一水平线上，滑动滚轴，你会发现发达国家与发展中国家的差距。而造成世界各国发展水平的差异我认为有以下几点原因：

- **劳动力的供给**。发达国家会吸引更多的劳动力流入，而发展中国家由于社会福利，薪资酬劳等问题会流失一些劳动力。
- **资本的供给**。发达国家的资本雄厚，有更多更充足的社会资源，而发展中国家由于国家正在发展，所以资本较少，产业都在起步阶段。
- **土地、矿产、石油等自然资源供给**。发达国家由于有资本，即使它的自然资源较少，也可以购买非发达国家的资源，让自己的产业得以发展。而发展中国家由于缺少相应的技术，只能卖最基本的能源获取利润。
- **技术**。发达国家各种高精尖技术已经日益成熟，而发展中国家大部分工作还需要人力资源，需要青壮劳动力。这也是为什么他们的生育率一直较高的缘故，所以世界各国GDP的数据是为少儿抚养比做铺垫的。

### 分析二
少年人口抚养比也称少年儿童抚养系数。指某一人口中少年儿童人口数与劳动年龄人口数之比。通常用百分比表示。以反映每100名劳动年龄人口要负担多少名少年儿童。计算公式为：（0-14岁少年儿童人口数 ∕ 劳动年龄人口数）※100%
由少儿抚养比的世界地图来看，基本上颜色的深浅分布与GDP恰恰相反。所以初步假设，国家的GDP与少儿抚养比成反比。

```python
from pyecharts import options as opts
from pyecharts.charts import Pie


def pie_base() -> Pie:
    c = (
        Pie()
        .add("", [list(b) for b in zip(list(df.sn),list(df.money))])
        .set_global_opts(title_opts=opts.TitleOpts(title=""))
        .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
    )
    return c
pie_base().render_notebook()
```

```python

from pyecharts import options as opts
from pyecharts.charts import Bar, Tab, Pie, Line

def bar_datazoom_slider() -> Bar:
    c = (
        Bar()
        .add_xaxis(list(df.sn))
        .add_yaxis("GDP",list(df.money))
        .add_yaxis("dependency",list(df.dependency),is_selected=False)
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Bar-DataZoom（slider-水平）"),
            datazoom_opts=[opts.DataZoomOpts()],
        )
    )
    return c

def line_markpoint() -> Line:
    c = (
        Line()
        .add_xaxis(list(df.sn))
        .add_yaxis(
            "GDP",
            list(df.money),
            markpoint_opts=opts.MarkPointOpts(data=[opts.MarkPointItem(type_="min")]),
        )
        .add_yaxis(
            "dependency",
            list(df.dependency),
            markpoint_opts=opts.MarkPointOpts(data=[opts.MarkPointItem(type_="max")]),
        )
        .set_global_opts(title_opts=opts.TitleOpts(title="Line-MarkPoint"))
    )
    return c

tab = Tab()
tab.add(bar_datazoom_slider(), "bar-example")
tab.add(line_markpoint(), "line-example")
tab.render_notebook()
```

### 总结
为了更方便对比，笔者在发达国家和发展中国家这两大分类中进行数据抽取，接着用抽取的国家和儿童抚养比画了一个漏斗图，用抽取的国家和国家的GDP画了一个饼图。最后把所有数据综合，画出了一个组合图表。
综合所有的图表信息，我们可以发现世界各国的儿童抚养比和世界各国的GDP两者差异非常大。
发达国家GDP高，少儿抚养比低；发展中国家GDP低，少儿抚养比低。所以我的结论是国家的经济发展水平与国家少儿抚养比成反比。假设成立。

**我认为造成这个结论的原因如下**：
- 发达国家经济发展水平高，社会福利制度好。因为商品经济高度发达，社会福利建设完善，所以国家社会福利制度好。国家居民生活水平高，人们没有传统的“养儿防老”这种思想。因此生育率自然下降，所以少儿抚养比也会下降。
- 各种高新技术产业发达。参考这篇文章。[从美国新经济看中国高新技术产业的发展策略](https://wenku.baidu.com/view/6861ba6c4bfe04a1b0717fd5360cba1aa8118cff.html)

以美国为例，高新技术产业的发展是美国经济持续出长的根本动力。高新技术产业已成为美国经济的“支柱产业”。以信息产业为例，从1990年一1996年，美国信息产业的销售额增长了57％，达到8600亿美元。信息产业已超过美国80年代的三大支柱产业：汽车业、建筑业、钢铁业，而一跃成为美国经济的头号支柱产业。在1993年一1997年间，高科技部门产值在美国GDP中所占的比重已达27％。据美国经济学家预测，到下世纪初，高科技部门将占美国GDP的近一半。高新技术产业发达，各种人工智能已经逐渐代替简单而枯燥的人类工作。所以相对应的发达国家所需要的青壮劳动力并没有发展中国家多。劳动力需求较小，所以生育率下降，少儿抚养比较低。
- 文化思想观念差异。发达国家社会制度建立完善，十分重视教育的普及。义务教育，高等教育等制度法规有明确的规定。普遍的国民素质较高，对于生育观也有自己的看法。但是发展中国家的普遍的思想是“多子多福”，生育的多，家里的劳动力也会多，因为生育率高居不下，少儿抚养比也会高。
- 国家政策差异。少儿抚养比中的支出很大一部分是在儿童的教育上，其次才是吃穿喝等等需求。但是发达国家的社会福利制度好，国家的经济支出很大一部分给了教育，给了许多教育的政策支持。因此，国民对教育的投入会少。但是发展中国家相对没有那么注重教育。所以对教育的扶持政策不够大，教育这项支出就在少儿抚养占比较高。
